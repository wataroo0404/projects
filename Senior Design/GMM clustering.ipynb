```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame = pd.read_csv('4-7_cluster.csv', index_col='division')
#data[['Business Unit','Latitude','Longitude']].groupby(['Business Unit'])
data.head(), data.shape
```

```python
from sklearn.mixture import GaussianMixture
train = data.reset_index(drop=True).values
gmm = GaussianMixture(n_components=17).fit(train)
labels = gmm.predict(train)
plt.scatter(train[:, 0], train[:, 1], c=labels, s=100, cmap='viridis');
```
